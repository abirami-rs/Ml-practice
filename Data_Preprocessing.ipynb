Feature scaling (Standardization, Normalization)



```python
#Normalization(Min-max scaling)
from sklearn.preprocessing import MinMaxScaler
scaling=MinMaxScaler()
X=[column for column in df.columns if df[column].dtype!=object]
minmax_scaling=scaling.fit_transform(df[X])
print(minmax_scaling)

#1   3   5   7
#After Min-Max scaling:
#0   0.33   0.66   1
#1 → 0 (minimum)
#7 → 1 (maximum)
#3 and 5 are placed proportionally in between
```

```python
#StandardScaling
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
X=[col for col in df.columns if df[col].dtype!=object]
standard_scaling=scaler.fit_transform(df[X])
print(standard_scaling)
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Convert back to DataFrame for easier plotting
standard_df = pd.DataFrame(standard_scaling, columns=X)

# Plot each column
for col in standard_df.columns:
    plt.plot(standard_df[col], label=col)

plt.xlabel('Index')
plt.ylabel('Standardized Values')
plt.title('Standardized Data')
plt.legend()
plt.show()

```
